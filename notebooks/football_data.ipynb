```python
import requests
from pathlib import Path
from dotenv import load_dotenv
import json
import os
```

```python
dotenv_path = Path("/Users/arthurgonzaga/DataKick/.env")
load_dotenv(dotenv_path=dotenv_path)
api_key = os.getenv('KEY_FOOTBALL_DATA_PROD')
```

```python
# uri = 'https://api.football-data.org/v4/competitions/CLI/' ### Info do campeonato
uri = 'https://api.football-data.org/v4/competitions/CLI/teams?season=2024'
headers = { 'X-Auth-Token': api_key }
response = requests.get(uri, headers=headers)
res_json = response.json()
```

```python
with open("/Users/arthurgonzaga/DataKick/notebooks/temp/football_data/times_libertadores.json", "w") as json_file:
    json.dump(res_json, json_file, indent=4)
```

```python
for teams in res_json['teams']:
    print(teams['name'])
    team_name = teams['name']
    
    dir_path = f"/Users/arthurgonzaga/DataKick/notebooks/temp/football_data/{team_name}"
    file_path = os.path.join(dir_path, "squad.json")
    
    os.makedirs(dir_path, exist_ok=True)

    with open(file_path, "w") as json_file:
        json.dump(teams['squad'], json_file, indent=4)
```

```python
for teams in res_json['teams'][:1]:
    print(teams['name'])
    team_name = teams['name']
    
    dir_path = f"/Users/arthurgonzaga/DataKick/notebooks/temp/football_data/{team_name}"
    file_path = os.path.join(dir_path, "squad.json")

    with open(file_path, "r") as json_file:
        squad = json.load(json_file)

    for player in squad[:1]:
        player_id = player['id']
        uri_base = f'https://api.football-data.org/v4/persons/{player_id}/matches?competitions=CLI&dateFrom=2024-01-01&dateTo=2024-12-31'
        print(player_id, uri_base)
        headers = { 'X-Auth-Token': api_key }
        response = requests.get(uri_base, headers=headers)
        player_json = response.json()
```
